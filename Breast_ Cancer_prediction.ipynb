```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('data_breast_cancer.csv')
```

```python
df
```

```python
df.info()
```

```python
df.diagnosis.value_counts()
```

```python
df.drop(columns=['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
df.head(5)
```

```python

plt.style.use('fivethirtyeight')
# for interactive visualizations
```

```python
df.describe()
```

```python
df['diagnosis']=df['diagnosis'].map({'M':1,'B':0})
```

```python
df.head(5)
```

```python
df.diagnosis.value_counts()
```

```python
x=df.drop('diagnosis',axis=1)
y=df['diagnosis'].values
```

```python
#scaling features 
# note that the scale function standardise  each columns
 #x=x-mean/std(x)
from sklearn.preprocessing import scale
x=scale(x)
```

```python
x
```

```python
from sklearn. model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,train_size=0.7,test_size=0.3,random_state=42)
```

```python
from sklearn.svm import SVC
model= SVC(C=1)
new_model=model.fit(train_x,train_y)
```

```python
y_pred=new_model.predict(test_x)
```

```python
y_pred
```

```python
from sklearn import metrics
metrics.confusion_matrix(y_true=test_y,y_pred=y_pred)
```

```python
print('accuracr:',metrics.accuracy_score(test_y,y_pred))
```

```python
print('precission:',metrics.precision_score(test_y,y_pred))
```

```python
print('recall:',metrics.recall_score(test_y,y_pred))
```

```python
#cross validation using k-fold
from sklearn.model_selection import validation_curve
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
folds=KFold(n_splits=4,shuffle=True,random_state=4)
model=SVC(C=1)
csv_result=cross_val_score(model,train_x,train_y,cv=folds,scoring='accuracy')
```

```python
print(csv_result)
```

```python
print(csv_result.mean())
```

```python
#grid search to find optimal Hyperperameter C
#spacify range of parameter (c) as a list
params={'C':[0.1,0.5,1,1.5,10,100,100]}
model=SVC()
#set up gridsearch scheme
# we are using kfold 5 as we take it earlier 
model_cv=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=folds,verbose=1,return_train_score=True)
```

```python
model_cv.fit(train_x,train_y)
```

```python
cv_results=pd.DataFrame(model_cv.cv_results_)
cv_results
```

```python
plt.figure(figsize=(8,6))
plt.plot(cv_results['param_C'],cv_results['mean_test_score'])
plt.plot(cv_results['param_C'],cv_results['mean_train_score'])
plt.xlabel('C')
plt.ylabel('accuracy')
plt.legend(['test_accuracy','train_accuracy'],loc= 'upper left')
plt.xscale('log')
```

```python
#finally we look at the optimal value for c
best_score=model_cv.best_score_
best_c=model_cv.best_params_['C']
print(best_score)
```

```python
print(best_c)
```

```python
model=SVC(C=best_c)
model.fit(train_x,train_y)
y_pred=model.predict(test_x)
y_pred
print('accuracy:',metrics.accuracy_score(test_y,y_pred))
print('precission:',metrics.precision_score(test_y,y_pred))
print('recall:',metrics.recall_score(test_y,y_pred))
metrics.confusion_matrix(y_true=test_y,y_pred=y_pred)
```

```python
#improving my model
parm_grid={'C':[0.1,1,7,8,100],'gamma':[1,0.1,0.01,0.001,0.0001],'kernel':['rbf']}
```

```python
gridd=GridSearchCV(SVC(),parm_grid,refit=True,verbose=4)
gridd.fit(train_x,train_y)
```

```python
gridd.best_params_
```

```python
gridd_pre=gridd.predict(test_x)
```

```python
gridd_pre
```

```python
print('accuracy:',metrics.accuracy_score(test_y,gridd_pre))
```

```python
metrics.confusion_matrix(y_true=test_y,y_pred=gridd_pre)
```

```python
print('precission:',metrics.precision_score(test_y,gridd_pre))
print('recall:',metrics.recall_score(test_y,gridd_pre))
```
